Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Also mention any two industries where Tableau is commonly used and explain one use case.
--
    Tableau is a powerful Business Intelligence (BI) and data visualization tool used to analyze data and create interactive dashboards, charts, and reports. It helps users understand large amounts of data quickly through visual representations.

**Why Tableau is Preferred Over Traditional Spreadsheet Tools**
1. Interactive Visualizations
Tableau creates dynamic charts, graphs, maps, and dashboards that are more engaging and easier to understand than spreadsheet tables.
2. Handles Large Datasets
It can process and analyze millions of rows of data efficiently, whereas spreadsheet tools may become slow with large datasets.
3. Drag-and-Drop Interface
Users can build visualizations without extensive programming knowledge, making analysis faster and easier.
4. Real-Time Data Analysis
Tableau can connect directly to databases and update dashboards automatically when data changes.
5. Better Data Insights
Advanced analytics features help identify trends, patterns, and outliers more effectively than traditional spreadsheets.
** Industries Where Tableau is Commonly Used**
1. Healthcare
2. Retail
*Example Use Case (Retail Industry)*

A retail company uses Tableau to analyze sales data from different stores. Managers can view interactive dashboards showing:

1. Best-selling products
2. Regional sales performance
3. Customer purchasing trends
4. Inventory levels

This helps the company make data-driven decisions, optimize stock levels, and improve overall sales performance.

Q9.  What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)
--
Tableau Prep is a data preparation tool that helps users clean, combine, and organize data before analyzing it in Tableau. It provides a visual and user-friendly interface for preparing data without requiring advanced programming skills.

*Importance in Real-World Data Projects*
1. Improves Data Quality
    Cleans inaccurate, incomplete, or inconsistent data before analysis.
2. Saves Time
    Automates repetitive data cleaning tasks and reduces manual effort.
3. Combines Multiple Data Sources
    Allows data from different files, databases, and systems to be merged easily.
4. Ensures Accurate Insights
    Clean and well-structured data leads to more reliable reports and dashboards.
5. Supports Data Governance
    Creates repeatable data preparation workflows that can be reused and shared.
**How to Remove Null Values**
Steps:
1. Open the dataset in Tableau Prep.
2. Identify fields containing null values.
3. Click on the profile pane of the field.
4. Select the Null values.
5. Choose Exclude to remove rows containing null values.
6. Verify that the null records have been removed.
**How to Remove Duplicate Records**
Steps:
1. Load the dataset into Tableau Prep.
2. Add a Clean Step.
3. Review records for duplicate entries.
4. Use Group and Replace or an Aggregate Step to identify duplicates.
5. Keep only one unique record for each duplicate group.
6. Validate the output to ensure duplicates are removed.
**How to Change Incorrect Data Types**
Steps:
1. Open the dataset in Tableau Prep.
2. Locate the field with the incorrect data type.
3. Click the data type icon beside the field name.
4. Select the correct data type:
    String (Text)
    Number
    Date
    Date & Time
    Boolean
Verify that the values are displayed correctly after conversion.
Continue with further data preparation if required.
